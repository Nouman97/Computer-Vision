```
import requests

url = 'https://app.nanonets.com/api/v2/ObjectDetection/Model/91feac22-3541-42a4-a74c-a15a8aacc926/LabelFile/'

data = {'file': open('/content/gdrive/My Drive/dataset-resized/glass/glass103.jpg', 'rb')}

response = requests.post(url, auth=requests.auth.HTTPBasicAuth('4lJfob3n6oBRen5AVvsBTN7llzM6wkCB', ''), files=data)
```

```
print(response.text)
a = response.text.find("xmin")
b = response.text.find("ymin")
c = response.text.find("xmax")
d = response.text.find("ymax")
e = response.text.find("score")
print(response.text[a+6:b-2])
print(response.text[b+6:c-2])
print(response.text[c+6:d-2])
print(response.text[d+6:e-2])

f = (response.text[a+6:b-2])
g = (response.text[b+6:c-2])
h = (response.text[c+6:d-2])
i = (response.text[d+6:e-2])

import cv2
import matplotlib.pyplot as plt

j = cv2.imread('/content/gdrive/My Drive/dataset-resized/glass/glass103.jpg')
k = cv2.cvtColor(j, cv2.COLOR_BGR2RGB)

k = cv2.rectangle(k, (int(f), int(g)), (int(h), int(i)), (255, 0, 0), 5)
plt.imshow(k)  
```
