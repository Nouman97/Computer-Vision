```
# https://www.youtube.com/watch?v=PtCQH93GucA

import cv2
import numpy as np
import matplotlib.pyplot as plt
# to load data from google drive, we can use this command set

from google.colab import drive
drive.mount('/content/gdrive')
```

```
a = cv2.imread("/content/gdrive/My Drive/1.jpg")
b = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
cv2.circle(b, (950, 1000), 50, (255, 0, 0), -1)
cv2.circle(b, (1500, 1000), 50, (255, 0, 0), -1)
cv2.circle(b, (850, 1750), 50, (255, 0, 0), -1)
cv2.circle(b, (1600, 1750), 50, (255, 0, 0), -1)
plt.imshow(b)

pts1 = np.float32([[900, 1000], [1500, 1000], [850, 1750], [1500, 1750]])
pts2 = np.float32([[850, 1000], [1650, 1000],[850, 1850], [1600, 1850]])

# getting the perspective transform matrix
matrix = cv2.getPerspectiveTransform(pts1, pts2)

c = cv2.warpPerspective(b, matrix, (2000, 3000))

plt.figure()
plt.imshow(c)
```

```
image_1 = cv2.imread('/content/gdrive/My Drive/2.jpeg')
image = cv2.resize(image_1, (3000, 4000))
c_1 = cv2.warpPerspective(image, matrix, (3000, 4000))

plt.figure()
plt.imshow(image_1)
plt.figure()
plt.imshow(c_1)
```
